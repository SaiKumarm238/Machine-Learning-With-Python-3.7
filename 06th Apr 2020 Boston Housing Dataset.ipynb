Housing Values in Suburbs of Boston
The medv variable is the target variable.
Data description
The Boston data frame has 506 rows and 14 columns.
This data frame contains the following columns:
- crim- per capita crime rate by town.
- zn- proportion of residential land zoned for lots over 25,000 sq.ft.
- indus
proportion of non-retail business acres per town.
- chas -
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).
- nox - 
nitrogen oxides concentration (parts per 10 million).
- rm - 
average number of rooms per dwelling.
- age - 
proportion of owner-occupied units built prior to 1940.
- dis - 
weighted mean of distances to five Boston employment centres.
- rad - 
index of accessibility to radial highways.
- tax - 
full-value property-tax rate per \$10,000.
- ptratio - 
pupil-teacher ratio by town.
- black - 
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.
- lstat - 
lower status of the population (percent).
- medv - 
median value of owner-occupied homes in \$1000s.

# Apply the Linear regression for entire data

```python
features = df.drop("medv", axis = 1)
target = df["medv"]
```

```python
# Divide the data into tarining and testing for all the below problems
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(features,target)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# Apply the Linear regression for entire data
from sklearn.linear_model import LinearRegression
LR= LinearRegression()
LR.fit(x_train,y_train)
y_pred = LR.predict(x_test) 
y_pred
```

```python
# Check the mse and r2score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
mse = mean_squared_error(y_pred,y_test)
print("the mean_squared_error is: ", mse)
r2 = r2_score(y_pred,y_test)
print("the r2_score_error is: ", r2)
```

# Aplying the Ridge Regression 

```python
from sklearn.linear_model import Ridge
for i in range (0,10,2):
    RR = Ridge(alpha=(i))
    RR.fit(x_train,y_train)
    y_pred = RR.predict(x_test)
    mse = mean_squared_error(y_pred,y_test)
    print("alpha =",i,"the mean_squared_error is: ", mse)
    r2 = r2_score(y_pred,y_test)
    print("alpha =",i,"the r2_score_error is: ", r2)
```

```python
# fit testing and traing data
for i in range (1,10,2):
    RR = Ridge(alpha=(i))
    RR.fit(x_test,y_test)
    y_pred = RR.predict(x_train)
    mse = mean_squared_error(y_pred,y_train)
    print("alpha =",i,"the mean_squared_error is: ", mse)
    r2 = r2_score(y_pred,y_train)
    print("alpha =",i,"the r2_score_error is: ", r2)
```

```python
from sklearn.model_selection import cross_val_score
for i in range(2,10):
    RR= Ridge()
    preds = cross_val_score(RR,features,target,cv =(i)) 
    print(i,"cross_val_score is", preds)
```
