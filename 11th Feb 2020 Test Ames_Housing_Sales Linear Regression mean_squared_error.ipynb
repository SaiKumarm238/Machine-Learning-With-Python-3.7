## Apply MinMaxScaler on the data

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
sc.fit(sq_num)
num_features_sc = pd.DataFrame(sc.transform(sq_num),columns  = sq_num.columns)
num_features_sc
```

- Create train and test splits of both data sets. To ensure the data gets split the same way, use the same random_state in each of the two splits.
- For the data set, fit a basic linear regression model on the training data.
- Calculate the mean squared error on both the train and test sets.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x_train,x_test,y_train,y_test = train_test_split(num_features_sc,target)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# for the test model
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred_lr = lr.predict(x_train)
print(y_pred_lr)
mse_train_lr = mean_squared_error(y_pred_lr,y_train)
mse_train_lr
```

```python
# for test the model
lr.fit(x_test,y_test)
y_pred_test_lr = lr.predict(x_test)
mse_test_lr =mean_squared_error(y_pred_test_lr,y_test)
mse_test_lr
```

### For the data set, fit KNeighnorRegressor model on the training data. Calculate the mean squared error on both the train and test sets.

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn= KNeighborsRegressor()
knn.fit(x_train,y_train)
y_pred_knn_train = knn.predict(x_train)
mse_knn_train = mean_squared_error(y_pred_knn_train,y_train)
mse_knn_train
```

```python
knn.fit(x_test,y_test)
y_pred_knn_test = knn.predict(x_test)
mse_knn_test =mean_squared_error(y_pred_knn_test,y_test)
mse_knn_test
```

#### - For The data set, Fit KNeighnorRegressor model on the training data with neighbors ranging from 2 to 15 with interval 2.
#### - Create a dictionary with keys as neighbors numbers. Values as list where first element in the list is mean squared error for train data, second element in the list is mean squared error for train data

```python
k_errors = {}
for i in range (2,15,2):
    knn = KNeighborsRegressor(n_neighbors= i)
    knn.fit(x_train,y_train)
    y_out_knn_ne_train = knn.predict(x_train)
    k_errors[i]= [mean_squared_error(y_out_knn_ne_train,y_train)]
    y_out_knn_ne_test = knn.predict(x_test)
    k_errors[i].append(mean_squared_error(y_out_knn_ne_test,y_test))
k_errors
```

### Plot a curve by taking neighbors on X axis and Error on Y axis.

- Plot the curve for training data error
- Plot the curve test data error

```python
plt.plot(list(k_errors.keys()),[x[0] for x in k_errors.values()],'r')
plt.plot(list(k_errors.keys()),[x[1] for x in k_errors.values()],'b')
plt.xticks(range(2,15))
```
